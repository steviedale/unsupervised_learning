```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('results/wine_quality/GaussianMixture_metrics.csv')
# df = pd.read_csv('results/mnist/GaussianMixture_metrics.csv')
```

```python
df.sort_values(by='num_components', inplace=True)
```

```python
plt.figure(figsize=(10, 8))
# metric = 'calinski_harabasz_score'          # higher is better
# metric = 'davies_bouldin_score'           # lower is better
metric = 'adjusted_rand_score'            # higher is better
# metric = 'adjusted_mutual_info_score'     # higher is better
# metric = 'homogeneity_score'              # higher is better
# metric = 'completeness_score'             # higher is better
# metric = 'silhouette_score'               # higher is better
plt.plot(df['num_components'], df[f'{metric}_mean'], label=metric.upper())
plt.fill_between(
    np.array(df['num_components']), 
    np.array(df[f'{metric}_mean'])-np.array(df[f'{metric}_std']), 
    np.array(df[f'{metric}_mean'])+np.array(df[f'{metric}_std']), 
    alpha=0.15)
plt.xlabel('Number of components')
metric_str = metric.replace('_', ' ').capitalize()
plt.ylabel(metric_str)
plt.title(f'{metric_str} vs Number of components')
```
