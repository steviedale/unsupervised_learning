```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from data_helpers.wine_quality_data_helper import load_wine_quality_data
from data_helpers.mnist_data_helper import load_mnist_data
```

```python
# load wine quality data
X_train, y_train, X_test, y_test = load_wine_quality_data()
```

```python
# concat y_train and y_test
wine_y = np.concatenate([y_train, y_test])
```

```python
# get value counts
wine_vc = dict(pd.Series(wine_y).value_counts())
wine_vc = {k: v for k, v in sorted(wine_vc.items(), key=lambda item: item[0])}
wine_vc
```

```python
plt.figure(figsize=(5, 6))
fontsize = 14
plt.title('Wine Quality Label Distribution', fontsize=fontsize)
plt.ylabel('Count', fontsize=fontsize)
plt.xlabel('Quality Score', fontsize=fontsize)
plt.hist(wine_y, bins=range(3, 10), align='left', rwidth=0.8)
```

```python
# load wine quality data
X_train, y_train, X_test, y_test = load_mnist_data()
```

```python
# concat y_train and y_test
mnist_y = np.concatenate([y_train, y_test])
```

```python
mnist_y.shape
```

```python
# get value counts
mnist_vc = dict(pd.Series(mnist_y).value_counts())
mnist_vc = {k: v for k, v in sorted(mnist_vc.items(), key=lambda item: item[0])}
mnist_vc
```

```python
plt.figure(figsize=(5, 6))
fontsize = 14
plt.title('MNIST Label Distribution', fontsize=fontsize)
plt.ylabel('Count', fontsize=fontsize)
plt.xlabel('Digit', fontsize=fontsize)
plt.bar(mnist_vc.keys(), mnist_vc.values(), align='center', color='green')
plt.xticks([0, 1, 2])
```

```python
# plot both distributions in one figure, left to right
figure, axis = plt.subplots(1, 2, figsize=(10, 3.5)) 

fontsize = 16

axis[0].set_title('MNIST Label Distribution', fontsize=fontsize)
axis[0].set_ylabel('Count', fontsize=fontsize)
axis[0].set_xlabel('Digit', fontsize=fontsize)
axis[0].bar(mnist_vc.keys(), mnist_vc.values(), align='center', color='green')
axis[0].set_xticks([0, 1, 2])

axis[1].set_title('Wine Quality Label Distribution', fontsize=fontsize)
axis[1].set_ylabel('Count', fontsize=fontsize)
axis[1].set_xlabel('Quality Score', fontsize=fontsize)
axis[1].hist(wine_y, bins=range(3, 10), align='left', rwidth=0.8)

# increase font of axis labels and ticks
axis[0].tick_params(axis='both', which='major', labelsize=fontsize-2)
axis[1].tick_params(axis='both', which='major', labelsize=fontsize-2)

# add spacing between plots
plt.tight_layout()

path = '/Users/sdale/repos/gatech/machine_learning/unsupervised_learning/final_results/datasets/label_distribution.png'
os.makedirs(os.path.dirname(path), exist_ok=True)
figure.savefig(path, dpi=300)
```
